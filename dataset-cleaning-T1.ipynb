```python
import pandas as pd
```

Read datasetOne, see what columns there are and what column I'll need for these tasks.

```python
dataset_one = pd.read_csv('datasetOne.csv')
dataset_one
```

```python
dataset_one.columns
```

Choose the column I needed and make all text lower case to keep all data cohesive.

```python
dataset_one['Food_items'] = dataset_one['Food_items'].str.lower()
```

```python
dataset_one_subset = dataset_one['Food_items']
dataset_one_subset
```

Drop any reaccuring items to ensure there are no multiples.

```python
dataset_one_subset = dataset_one_subset.drop_duplicates()
dataset_one_subset
```

Repeat this basic reading of the datasets and cleaning with the other CSV's I downloaded from [Kaggle](https://www.kaggle.com/datasets/dtamindu/food-items-with-nutritional-info)

```python
dataset_two = pd.read_csv('datasetTwo.csv')
dataset_two
```

```python
dataset_two_subset = dataset_two['food']
dataset_two_subset
```

```python
dataset_two_subset = dataset_two_subset.drop_duplicates()
dataset_two_subset
```

```python
dataset_three = pd.read_csv('datasetThree.csv')
dataset_three
```

```python
dataset_three['Food'] = dataset_three['Food'].str.lower()
```

```python
dataset_three_subset = dataset_three['Food']
dataset_three_subset
```

```python
dataset_three_subset = dataset_three_subset.drop_duplicates()
dataset_three_subset
```

```python
dataset_four = pd.read_csv("datasetFour.csv", on_bad_lines="skip")
dataset_four
```

```python
dataset_four.columns
```

```python
dataset_four_subset = dataset_four['Food']
dataset_four_subset
```

```python
dataset_four_subset = dataset_four_subset.drop_duplicates()
dataset_four_subset
```

Use concatenation to take all the columns and join them together as one big column. A big food dataset like this will hopefully allow for clearer and more accurate results when training the models.

```python
all_foods = pd.concat([
    dataset_one_subset,
    dataset_two_subset,
    dataset_three_subset,
    dataset_four_subset
], ignore_index=True)
```

Clean again like earlier to ensure everything is cohesive.

```python
all_foods = all_foods.str.lower()
```

```python
all_foods = all_foods.drop_duplicates()
```

Name column.

```python
all_foods_df = pd.DataFrame({'Food': all_foods})
```

```python
all_foods_df
```

Save to CSV.

```python
all_foods_df.to_csv('Food.csv', index=False)
```
