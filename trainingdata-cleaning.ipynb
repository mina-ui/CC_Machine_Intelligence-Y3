```python
import pandas as pd
```

```python
training_data = pd.read_csv("foodTrainingSubset.csv", header=0)
training_data
```

```python
training_data.columns
```

Making everything a string type and all in lower case.

```python
training_data["Food"] = training_data["Food"].astype(str).str.lower()
training_data["Categories"] = training_data["Categories"].astype(str).str.lower()
training_data
```

Make Categories back to obj after putting them in lowercase (ChatGPT)

```python
training_data["Categories"].replace("nan", pd.NA, inplace=True)
```

Final cleaning, removing some fast-food terms, any empty rows, duplicates, whitespaces etc.

```python
remove_words = [
    "cooked", "stewed", "canned", "arby's", "burger king", "mcdonalds",
    "mcdonald's", "mcflurry", "wendy's", "jr.", "zinger", "whopper",
    "subway", "pizza hut"
]

def clean_item(name):
    for w in remove_words:
        name = name.replace(w, "")
    return name.strip()

training_data["Food"] = training_data["Food"].apply(clean_item)

# remove empty rows (ChatGPT)
training_data = training_data[training_data["Food"].str.len() > 0]
```

```python
training_data = training_data.drop_duplicates(subset="Food").reset_index(drop=True)
```

```python
print(training_data)
```

Dropping any empty rows and saving to CSV.

```python
labeled_data = training_data.dropna(subset=["Categories"])
labeled_data
```

```python
labeled_data.to_csv('labeledData3.csv', index=False)
```
